```python
import cv2
from matplotlib import pyplot as plt

# Читать изображение
img = cv2.imread("image.jpg")  # путь к твоему файлу

# Конвертация BGR → RGB
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

# Показать изображение
plt.imshow(img_rgb)
plt.axis('off')
plt.show()
```
