Object Detection отвечает на 2 вопроса одновременно:

Что находится на изображении?

Где это находится?

Разница между задачами CV
Задача	Результат
Classification	один класс
Detection	класс + bounding box
Segmentation	маска объекта
Bounding Box — ключевой объект detection

Форматы координат:

x_min, y_min, x_max, y_max

или

x_center, y_center, width, height

⏱️ Блок 2 (15 минут) — Главная метрика: IoU

Почему модель нельзя понять без метрик.

IoU = насколько bbox совпал с truth

IoU = \frac{Area\ of\ Overlap}{Area\ of\ Union}

Что важно понять

IoU > 0.5 → good detection

IoU используется в:

evaluation

NMS

mAP

⏱️ Блок 4 (20 минут) — Как работает YOLO внутри

Понять YOLO = понять современный detection.

YOLO идея:

Изображение делится на grid.

Каждая клетка предсказывает:

bbox

confidence

class

confidence =
objectness × class probability
Почему нужен NMS

Модель выдаёт много overlapping boxes.

NMS удаляет лишние.
